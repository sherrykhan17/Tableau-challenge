```python
# NewYork Citibike Program 2020 -  Oct - Dec trip data
import pandas as pd
import datetime
import numpy as np
import os
```

```python
# Get NewYork Citibike tripdata csv files from Oct - Dec 2020
filepath_oct = os.path.join("Citi Bike Data" ,"JC-202010-citibike-tripdata.csv")
filepath_nov = os.path.join("Citi Bike Data" ,"JC-202011-citibike-tripdata.csv")
filepath_dec = os.path.join("Citi Bike Data" ,"JC-202012-citibike-tripdata.csv")

nycbike_2020oct = pd.read_csv(filepath_oct)
nycbike_2020nov = pd.read_csv(filepath_nov)
nycbike_2020dec = pd.read_csv(filepath_dec)
```

```python
nycbike_2020oct.head(5)
```

```python
nycbike_last_quater_2020 = pd.concat([nycbike_2020oct,nycbike_2020nov,nycbike_2020dec])
```

```python
# Output from Oct - Dec 2020 citibike tripdata to CSV file 
filepath_2020_last_quater = os.path.join("Citi Bike Data" ,"JC-2020_last_quater-citibike-tripdata.csv")
nycbike_last_quater_2020.to_csv(filepath_2020_last_quater, index=False)
nycbike_last_quater_2020.tail(5)
```
